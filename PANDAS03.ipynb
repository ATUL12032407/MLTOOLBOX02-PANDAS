Merge and join

```python
df_1=pd.DataFrame({'key1':[1,2,3,4,5],'key2':[6,7,8,9,4],'key3':[3,5,7,8,1]})
df_1
```

```python
df_2=pd.DataFrame({'key1':[13,2,34,4,51],'key4':[66,7,87,9,46],'key5':[13,50,75,18,1]})
df_2
```

```python
# merge and join are not same
```

```python
# merge happen on column
pd.merge(df_1,df_2)     #inner join(intersection)
```

```python
pd.merge(df_1,df_2,how='left')  # left join(all the elements from the left side and common elements of right side)
```

```python
df_1
```

```python
df_2
```

```python
pd.merge(df_1,df_2,how='right')   # right join
```

```python
pd.merge(df_1,df_2,how='outer')   # all the key use of both the data
```

```python
pd.merge(df_1,df_2,how='cross')
```
