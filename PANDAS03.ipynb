Merge and join

```python
df_1=pd.DataFrame({'key1':[1,2,4,5,6],'key2':[4,5,6,7,8],'key3':[3,4,5,6,6]})
df_1
```

```python
df_2=pd.DataFrame({'key1':[1,2,45,6,67],'key4':[56,5,6,7,8],'key5':[3,56,5,6,6]})
df_2
```

```python
# merge and join are not same
```

```python
# merge happen on column
pd.merge(df_1,df_2)     #inner join(intersection)
```

```python
pd.merge(df_1,df_2,how='left')  # left join(all the elements from the left side and common elements of right side)
```

```python
df_1
```

```python
df_2
```

```python
pd.merge(df_1,df_2)
```

```python
#Merge
pd.merge(df_1,df_2,how='left')
```

In left merge all the elements from the left side and common element of the right data side

```python
 # right join
pd.merge(df_1,df_2,how='right')
```

In right merge all the elements from the right side and common element of the left data side

```python
# all the key use of both the data
pd.merge(df_1,df_2,how='outer')
```

```python
df_1
```

```python
df_2
```

```python
pd.merge(df_1,df_2,how='cross')
```

```python
pd.merge(df_1,df_2,how='left')
```

```python
# If we don't want some specific common attribute as a key element.

pd.merge(df_1,df_2,how='left',left_on = 'key2',right_on ='key4')
```

```python
df_1.join(df_2, how='left', lsuffix='_df1', rsuffix='_df2')
#Join happens on the indexes , column name should not be found.
```
