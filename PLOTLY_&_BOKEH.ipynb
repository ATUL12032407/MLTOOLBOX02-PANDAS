```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = [1, 2, 3, 4, 5], y = [3, 4, 5, 6, 7], mode = "markers"))
fig.show()
```

```python
#add_trace is method used to add one or more traces to a plot. Traces represent individual sets of data and other aspects of plot
#that can be visualised on the same plot.
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[3,4,5,6,7],mode ="lines"))
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[3,4,5,6,7],mode ="lines+markers"))
fig.show()
```

```python
# What is size of people who came for dinner ?
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill , y =tips.tip ,mode ="markers", marker_size =tips['size']))
fig.show()
```

```python
# What is size of people who came for dinner ?
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill , y =tips.tip ,mode ="markers", marker_size =12*tips['size']))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill , y =tips.tip ,mode ="markers", marker_size =10*tips['size'] , z=tips['size']))
fig.show()
```

```python
from bokeh.plotting import figure ,output_file, show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')
p =figure(title ='flower test')
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
output_file('test.html')
p =figure(title ='flower test')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
output_file('test.html')
p =figure(title ='flower test')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.line(flowers['petal_length'],flowers['petal_width'])
show(p)
```
